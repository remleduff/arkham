```python
trials = 1000

def run_test(strat):
    results = pd.DataFrame([], columns=['difficulty', 'curses', 'blessings', 'success'])
    for difficulty in range(-2, 6):
        sim = run_sim(success_probability, strat, difficulty=-difficulty, trials=trials)
        f = pd.DataFrame(sim, columns=range(11))
        f['curses'] = f.index
        f['difficulty'] = difficulty
        f = f.melt(id_vars=['difficulty', 'curses'], var_name='blessings', value_name='success')
        
        sim = run_sim(best_pendulum_guess, strat, difficulty=-difficulty, trials=trials)
        g = pd.DataFrame(sim, columns=range(11))
        g['curses'] = g.index
        g['difficulty'] = difficulty
        g = g.melt(id_vars=['difficulty', 'curses'], var_name='blessings', value_name='pendulum')

        f['pendulum'] = g['pendulum']
        
        results = results.append(f)
    return results.pivot(index=['difficulty', 'curses'], columns='blessings')
```

```python
def jacqueline_token_strat(bag, difficulty=-2, **kwargs):
    tokens, remaining = pull_tokens(bag, 3)
    outcome = list(tokens)

    blurse = blursed(tokens)
    m = [t for t in tokens if t not in blurses]
    best = max(m, key=modifier) if m else TENTACLE
    if TENTACLE in tokens:
        # Jacqueline can only cancel the tentacle, which leaves the other two tokens
        outcome.remove(TENTACLE)
    elif modifier(best) >= difficulty:
        # We can pass the test, cancel the other tokens
        outcome = [best]
    elif blurse:
        # We can't pass the test, take a blessing or curse, preferring blessings, then resolve
        outcome = [blurse]
        tokens = [blurse]
        while blursed(tokens):
            tokens, remaining = pull_tokens(remaining, 1)
            for token in tokens:
                outcome.append(token)
    return outcome
```

# Jacqueline, then Olive

(resolve either a single token from the first two pulled or, the two best from the next 3)

```python
def olive(bag, difficulty=-2, **kwargs):
    tokens, remaining = pull_tokens(bag, 2)
    olive, remaining = pull_tokens(remaining, 3)

    best_non_blurse = best_token(tokens, ignore=blurses)
    best = best_token(tokens)

    if modifier(best_non_blurse) >= difficulty:
        return [best_non_blurse]
    
    worst_non_curse_olive = worst_token(olive, ignore=blurses)
    worst_olive = worst_token(olive)
    
    non_blurse_olive = [t for t in olive if t not in blurses]
    if len(non_blurse_olive) == 2 and modifier(non_blurse_olive) >= difficulty:
        return non_blurse_olive

    if modifier(worst_non_curse_olive) <= modifier(CURSE):
        olive.remove(worst_non_curse_olive)
    else:
        olive.remove(worst_olive)

    val_olive = modifier(olive)
    
    if not blursed(olive) and val_olive >= difficulty:
        return olive
    
    if blursed(olive) and val_olive >= modifier(best):
        tokens = olive
    else:
        tokens = [best]
     
    return resolve_blurses(tokens, remaining)
```

```python
run_test(olive)
```

# Olive first, then Jaqueline

(Resolve the two best tokens from 5 pulled)

```python
def modifier_prefer_blurse(token):
    if token == TENTACLE:
        return float('-inf')
    if token in spooky:
        return spooky[token] + 0.5
    return token

def olive2(bag, difficulty=-2, **kwargs):
    tokens, remaining = pull_tokens(bag, 5)
    
    tokens.sort(reverse=True, key=modifier_prefer_blurse)
    
    non_blursed_tokens = [t for t in tokens if t not in blurses]
    if len(non_blursed_tokens) >= 2 and modifier(non_blursed_tokens[:2]) >= difficulty:
        return non_blursed_tokens[:2]
    
    # Corner case, there might be a bunch of tokens that are "better" than curses at the head of the list
    #   But they might not sum to enough to pass the test, so we need to be willing to take curses instead
    if not blursed(tokens[:2]) and blursed(tokens) and modifier(tokens[:2]) < difficulty:
        blursed_tokens = [t for t in tokens if t in blurses]        
        n = min(len(blursed_tokens), 2)
        tokens[:n] = blursed_tokens[:n]
    
    return resolve_blurses(tokens[:2])
```

```python
run_test(olive2)
```
